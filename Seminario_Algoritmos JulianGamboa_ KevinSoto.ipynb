Optimización del código anterior

```python
def backtracking_optimizado(matriz_tomas, toma_actual=0, dias=None, estado=None, mejor_coste=None, coste_acumulado=0, actores_por_dia=None):
    if dias is None:
        dias = []
    if estado is None:
        estado = [0]
    if mejor_coste is None:
        mejor_coste = [9999]
    if actores_por_dia is None:
        actores_por_dia = [] # Lista de sets para rastrear actores por cada día abierto

    # PODA: Si el coste que llevamos ya es peor que el mejor encontrado, paramos esta rama
    if coste_acumulado >= mejor_coste[0]:
        return

    # SOLUCIÓN COMPLETA ya ha recorrido toda la rama y sigue siendo optima entonces guarda el plan
    if toma_actual == len(matriz_tomas):
        if coste_acumulado < mejor_coste[0]:
            mejor_coste[0] = coste_acumulado
            estado[0] += 1
            print(f"solucion: {estado[0]} | coste: {coste_acumulado} | plan: {dias}")
        return

    # Condición de Parada Temprana
    if estado[0] >= num_soluciones:
        return

    actores_toma_actual = set(actores_por_toma[toma_actual])

    # itera sobre los días
    for i in range(len(dias)):
        if len(dias[i]) < max_tomas:
            # Calculamos el coste actual
            actores_previos = actores_por_dia[i]
            nuevos_actores = actores_toma_actual - actores_previos
            incremento = len(nuevos_actores)

            # Backup para el backtrack
            dias[i].append(toma_actual)
            actores_por_dia[i] = actores_previos | actores_toma_actual
            
            backtracking_optimizado(matriz_tomas, toma_actual + 1, dias, estado, mejor_coste, 
                                    coste_acumulado + incremento, actores_por_dia)
            
            # Backtrack (deshacer cambios)
            dias[i].pop()
            actores_por_dia[i] = actores_previos
            
        # Si el día está vacío, no hace falta probar los siguientes días vacíos ya cuando terminá
        if len(dias[i]) == 0:
            break

    ## Crea un nuevo dia teniendo en cuenta el maximo de tomas
    if len(dias) < max_dias:
        dias.append([toma_actual])
        actores_por_dia.append(actores_toma_actual)
        
        # Al ser un día nuevo, el coste sube por todos los actores de la toma
        backtracking_optimizado(matriz_tomas, toma_actual + 1, dias, estado, mejor_coste, 
                                coste_acumulado + len(actores_toma_actual), actores_por_dia)
        
        dias.pop()
        actores_por_dia.pop()
```

### Complejidad de backtracking_optimizado

```python
def optimizar_rodaje(lista_actores_tomas, tomas_por_dia=6):
    tomas_bits = []
    ## Representación por Máscaras de Bits (Bitmasking)
    for actores in lista_actores_tomas:
        mask = 0
        for a in actores:
            mask |= (1 << a)
        tomas_bits.append(mask)

    num_tomas = len(tomas_bits)
    num_dias = math.ceil(num_tomas / tomas_por_dia)

    estado = {
        "mejor_costo": float('inf'),
        "mejor_plan": [],
        "nodos_explorados": 0,
        "podas": 0,
        "mejoras": 0,
    }
    memo = {}

    inicio_t = time.time()
    print(f"\n EJECUTANDO: ({num_tomas} tomas, {num_dias} días)")
    ## Función Heurística Admisible (Branch and Bound)
    def obtener_cota(mask_tomas, costo_actual):
        actores_pendientes = 0
        for i in range(num_tomas):
            if not (mask_tomas & (1 << i)):
                actores_pendientes |= tomas_bits[i]
        return costo_actual + bin(actores_pendientes).count('1')

    def backtrack_2(mask_tomas, costo_acum, plan, dia_actual):
        estado["nodos_explorados"] += 1

        if obtener_cota(mask_tomas, costo_acum) >= estado["mejor_costo"]:
            estado["podas"] += 1
            return
        ## Memoización (Programación Dinámica superpuesta)
        if mask_tomas in memo and memo[mask_tomas] <= costo_acum:
            estado["podas"] += 1
            return
        memo[mask_tomas] = costo_acum

        if mask_tomas == (1 << num_tomas) - 1:
            if costo_acum < estado["mejor_costo"]:
                estado["mejor_costo"] = costo_acum
                estado["mejor_plan"]  = plan
                estado["mejoras"]    += 1
                print(f" Nuevo óptimo parcial: {estado['mejor_costo']}")
            return

        faltan = [i for i in range(num_tomas) if not (mask_tomas & (1 << i))]
        max_tomas_hoy = min(tomas_por_dia, len(faltan))
        ## Restricción del Espacio de Búsqueda
        semilla    = faltan[0]
        candidatas = faltan[1:13]
        ## Generación de Días por Combinatoria Interna
        def generar_dia(idx, actual):
            if len(actual) == max_tomas_hoy - 1:
                grupo = [semilla] + actual
                m_dia_tomas  = mask_tomas
                m_actores_dia = 0
                for t in grupo:
                    m_dia_tomas   |= (1 << t)
                    m_actores_dia |= tomas_bits[t]
                costo_dia = bin(m_actores_dia).count('1')
                backtrack_2(
                    m_dia_tomas,
                    costo_acum + costo_dia,
                    plan + [grupo],
                    dia_actual + 1
                )
                return
            for i in range(idx, len(candidatas)):
                actual.append(candidatas[i])
                generar_dia(i + 1, actual)
                actual.pop()

        generar_dia(0, [])

    backtrack_2(0, 0, [], 0)

    fin_t       = time.time()
    tiempo_total = fin_t - inicio_t
    eficacia_poda = (
        (estado["podas"] / estado["nodos_explorados"]) * 100
        if estado["nodos_explorados"] > 0 else 0
    )

    print("\n===== METRICAS =====")
    print("Tiempo de ejecución:",   round(tiempo_total, 2), "s")
    print("Nodos explorados:",       estado["nodos_explorados"])
    print("Podas realizadas:",        estado["podas"])
    print("Mejoras de diversificación:", estado["mejoras"])
    print("Costo óptimo final:",      estado["mejor_costo"])
    print("Plan óptimo final:",       estado["mejor_plan"])
    print("Eficacia de la poda:",     round(eficacia_poda, 10), "%")

    return estado["mejor_costo"], estado["mejor_plan"]

costo_A, plan_A = optimizar_rodaje(actores_por_toma)
```

#### Analisis del algoritmo de Optimización de Rodaje.
El algoritmo desarrollado es un Branch and Bound (Ramificación y Poda) optimizado con Programación Dinámica. Su objetivo es encontrar la mejor organización de rodaje sin tener que probar todas las combinaciones posibles.

En lugar de manejar listas de nombres, usamos números binarios. Esto hace que el programa sea increíblemente rápido al comparar estados.
Para evitar probar el mismo plan pero en diferente orden, el algoritmo siempre obliga a que la primera toma disponible sea la "semilla" del día actual. Esto reduce drásticamente el espacio de búsqueda. Si el algoritmo llega a una combinación de tomas que ya evaluó antes (pero por un camino diferente), y el costo actual es mayor al guardado, deja de buscar por ahí.

El diseño del algoritmo se basa en un equilibrio técnico entre la exploración exhaustiva y la explotación de soluciones prometedoras: 

Diversificación (Exploración): Se garantiza mediante el esquema de Backtracking puro. El algoritmo no se detiene al encontrar la primera solución válida; gracias a la recursión, es capaz de retroceder y explorar regiones del espacio de búsqueda completamente distintas. Esto evita caer en "óptimos locales" y asegura que todo el espectro de combinaciones de tomas sea considerado.

Intensificación (Explotación): Se logra mediante el uso de una Ventana de Candidatas ($W=12$) y la Poda por Cota Inferior. Una vez que el algoritmo identifica una rama con un costo prometedor, concentra sus recursos en "pulir" esa solución. La poda elimina trillones de opciones mediocres, permitiendo que el tiempo de cómputo se dedique solo a las soluciones que realmente pueden mejorar el récord actual.

#### Diseño de la Cota Inferior.

El corazón de esta funcionalidad es la cota inferior, este componente que le agrega el toque heuristico al algoritmo se basa en el principio de Actores Pendientes. Calcula el número de actores únicos que aún no han filmado todas sus escenas. Matemáticamente, es imposible terminar el rodaje con un costo menor a:$$\text{Costo Actual} + \text{Conteo de Actores Únicos Restantes}$$

Este diseño garantiza que el algoritmo sea óptimo, ya que la cota nunca sobreestima el costo real (es optimista), permitiendo descartar ramas con la certeza matemática de que no contienen la solución ideal.
